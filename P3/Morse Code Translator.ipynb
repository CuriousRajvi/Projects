# Morse Code Translator

<ul>
    <li>Morse code is a system of communication that uses dots, dashes, and spaces to represent the letters of the alphabet, numbers, and punctuation.
    <li>Originally, telegraph operators used Morse code to transmit messages using electrical pulses that they then translated into text messages.
    <li>Later, radio communicators adapted the code and began to use it as an auditory language that comprises signals as easily translatable dots and dashes.
